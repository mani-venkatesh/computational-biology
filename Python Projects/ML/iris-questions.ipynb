Import Data and Create Dataframe

```python
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
species = ['setosa', 'versicolor', 'virginica']
species_target = [None]*len(iris.target)
for i, num in enumerate(iris.target):
    species_target[i] = species[num]
df['species'] = species_target
df.insert(loc=0, column="id", value=range(1,df.shape[0]+1), allow_duplicates=True)
df.head()
```
